### Importing TMDB dataset

```python
data = pd.read_csv("../datasets/raw/tmdb/tmdb_5000_movies.csv")
display(data.head())
print(data.columns)
print(data.shape)
```

### Feature Generator

#### Release date -> release year

```python
release_year = [year.split('-')[0] for year in data["release_date"].values]
data["release_year"] = release_year
data.drop(columns=["release_date"], inplace=True)
data
```

### List separator

```python
data_lists = data.loc[:, ["genres", "production_companies", "production_countries", "spoken_languages"]]
data_lists
```

```python
x = data.loc[19995, "genres"]
lst = x.strip('][').split("}, {")
genres = [name.split("name\": \"")[-1].split("\"")[0].lower().replace(" ", "_") for name in lst]
genres
```

```python
x = data.loc[19995, "production_companies"]
lst = x.strip('][').split("}, {")
companies = [name.split("name\": \"")[-1].split("\"")[0].lower().replace(" ", "_") for name in lst]
companies
```

```python
x = data.loc[19995, "production_countries"]
lst = x.strip('][').split("}, {")
countries = [name.split("iso_3166_1\": \"")[-1].split("\"")[0].lower().replace(" ", "_") for name in lst]
countries
```

```python
x = data.loc[19995, "spoken_languages"]
lst = x.strip('][').split("}, {")
lang = [name.split("iso_639_1\": \"")[-1].split("\"")[0].lower().replace(" ", "_") for name in lst]
lang
```

```python
values = data.loc[:, "genres"]
values_list = [
    list_.strip('][').split("}, {") for list_ in values
]
values_keys = [[name.split("name\": \"")[-1].split("\"")[0].lower().replace(" ", "_") for name in v] for v in values_list] 
values_keys
```

```python
data_lists["genres"] = values_keys
```
