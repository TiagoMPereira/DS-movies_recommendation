```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def min_max_scaler(serie):
    return np.round((serie -  serie.min()) / (serie.max() - serie.min()), 3)
```

```python
data = pd.read_csv("../datasets/movies_dataset_model.csv", index_col=0)
data_metadata = pd.read_csv("../datasets/movies_dataset_metadata.csv", index_col=0)
data_metadata.set_index("tmdb_id", inplace=True)
data.set_index("tmdb_id", inplace=True)
data
```

```python
data_model = data.copy()
```

```python
data_model["release_year"] = min_max_scaler(data_model["release_year"])
data_model["runtime"] = min_max_scaler(data_model["runtime"])
data_model["vote_average"] = min_max_scaler(data_model["vote_average"])
```

```python
X = data_model.values
idx = data_model.index
y = data_model.loc[88269, :].values.reshape(1, -1)
```

```python
y
```

```python
sim = cosine_similarity(X, y)
similarity = pd.Series(sim.reshape(1, -1)[0], idx)
```

```python
similarity.sort_values(ascending=False, inplace=True)
```
