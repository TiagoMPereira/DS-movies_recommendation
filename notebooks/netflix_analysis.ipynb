```python
import pandas as pd
```

```python
data = pd.read_excel("../datasets/raw/Netflix Dataset Latest 2021.xlsx")
data.head()
```

```python
data.columns
```

```python
data = data.loc[data["Series or Movie"] == "Movie"]
data
```

```python
columns_keep = ["IMDb Score", "Netflix Link", "IMDb Link", "Poster", "TMDb Trailer", "Genre"]
columns_drop = [col for col in data.columns if col not in columns_keep]
data.drop(columns=columns_drop, inplace=True)
```

```python
columns_drop
```

```python
data.drop(data.loc[data["IMDb Link"].isnull()].index, inplace=True)
data
```

```python
imdb_id = [id_.split("/")[-1] for id_ in data["IMDb Link"]]
data["IMDb id"] = imdb_id
```
